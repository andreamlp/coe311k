```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Taylor series approximation for sin(x)
def taylor_sin(x, n_terms):
    approximation = 0.0
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
        approximation += term
    return approximation

# x values from 0 to 10 with step 0.1
x_values = np.arange(0, 10.1, 0.1)

# Exact values using numpy
exact_values = np.sin(x_values)

# Store errors
errors = {}

# Compute approximations and absolute errors for 14 terms
for terms in [1, 2, 3, 4]:
    approx_values = np.array([taylor_sin(x, terms) for x in x_values])
    abs_error = np.abs(exact_values - approx_values)
    errors[terms] = abs_error

# Plot truncation error for each number of terms
for terms in [1, 2, 3, 4]:
    plt.figure()
    plt.plot(x_values, errors[terms])
    plt.xlabel("x")
    plt.ylabel("Absolute Error")
    plt.title(f"Truncation Error Using {terms} Term(s)")
    plt.show()
```

**Does the error decrease as more terms are included?**
    Yes, but mainly for small values of x. Near x = 0, each additional Taylor term significantly improves the approximation, and the error becomes extremely small.
    
**Does the error decrease uniformly for all values of x?**
    No. The error does not decrease uniformly. For larger values of x, the polynomial performs worse.
    
**Small vs. Large x**
    For small x: the series converges well and the error decreases rapidly.
    For larger x: the error increases rapidly.
