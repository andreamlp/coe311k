```python
import numpy as np
import math
import random

# Monte Carlo PI approximation

def approximatePi(number_of_shots):
    counter = 0
    for s in range (0, number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y <1):
            counter = counter + 1
    return 4 * counter/number_of_shots
```

```python
# Main code

sample_sizes = list(range(10, 1001, 20))
errors = []


true_pi = np.pi

for n in sample_sizes:
    calculated_pi = approximatePi(n)
    error = abs(true_pi - calculated_pi)
    errors.append(error)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(sample_sizes, errors, 'r--')
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Approximation of pi: Error vs. Number of Samples")
plt.grid(True)
plt.show()
```
