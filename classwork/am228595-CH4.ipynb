## Truncation Errors

Errors that arise when an infinite process (like a series or function evaluation) is approximated by a finite process.

This occurs when you truncate or approximate a mathematical
procedure, such as in numerical differentiation, integration, or solving
differential equations.

## In Class Challenge: Leibniz Series for pi

Code the Leibniz series for pi in Python.
Run through the first 5,000 terms of the series.

Every 100 terms, print out the relative error between the current
running calculated value of ùúã and numpy.pi
As each term is added to the summation, what is happening to our
error?

```python
import math
import numpy as np

running_sum = 0.0

for n in range(5000):
    term = ((-1)**n)/ (2*n +1)
    running_sum += term
    
    if (n+1) % 100 == 0:
        approx_pi = 4 * running_sum
        relative_error = abs((approx_pi - np.pi) / np.pi)
        print(f"After {n+1} terms: relative error = {relative_error}")
```
