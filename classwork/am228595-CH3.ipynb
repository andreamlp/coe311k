```python
import numpy as np
import math
import random

def approximatePi(number_of_shots):
    counter = 0
    for s in range (0, number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y <1):
            counter = counter + 1
    return 4 * counter/number_of_shots
```

```python
for i in range(0,100):
    calculated_pi = approximatePi(1000)
    true_pi = np.pi
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs((true_pi - calculated_pi)/true_pi)
    
    print(calculated_pi, true_error, relative_error)
```

```python
import matplotlib.pyplot as plt

plt.plot(true_error, 'r--')
plt.plot(relative_error, 'g-o')
plt.show()
```
