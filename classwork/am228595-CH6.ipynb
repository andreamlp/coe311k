Iterative methods start with a guess and refine it as it goes on.

**Convergence conditions:**
- matrix is diagonally dominant

**Advantages:**
- memory efficiency

**Disadvantages:**

**Recap:**

**Absolute Error:**
Definition: measures the raw difference between the true value and the approximation
- how far off are we?
- does not consider the scale of the true value
- You're working with values near 0/bounded by 0

**Relative Error:**
Definition: normalizes error by the size of the true value, sometimes expressed as a percentage.
- how big is the error compared to the true value?
- the scale does matter

**Jacobi Method:**
Key idea: Each variable is updated using only the previous values from the previous iteration.
Convergence: Slower but easy to implement?

**Tolerance:**
- defines how close the current solution must be to the true solution before stopping the interation.
- If the difference between consecutive iterations is smaller than the tolerance, the method is said to have converged.

```python
#JACOBI Method

#time module:
import time

import numpy as np

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x, dtype=float)
        
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
    
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new

    return x, max_iterations
```

```python
# Example usage-JACOBI

A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]])

b = np.array([14, 30, 26, 25, 37])

x0 = np.zeros_like(b)
tol = 1e-6

max_iterations = 100

# Start time
start_time = time.time()

solution, iterations = jacobi(A, b, x0, tol, max_iterations)

# End time
end_time = time.time()

#Elaspsed time
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time}")

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
```

```python
#GAUSS-SEIDEL Method

import numpy as np

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy().astype(float)
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k

        x = x_new

    return x, max_iterations
```

```python
# Example usage-GAUSS-SEILDEL 

A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]])

b = np.array([14, 30, 26, 25, 37])

x0 = np.zeros_like(b)
tol = 1e-6

max_iterations = 100

# Start time
start_time = time.time()

solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)

# End time
end_time = time.time()

#Elaspsed time
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time}")

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
```

**Conclusion**

- Jacobi and Gauss-Seidel methods are important iterative solvers for systems of linear equations, particularly when the matrix A is large and sparse.
- The Jacobi method is easier to understand but slower in convergence compared to Gauss-Seidel, which updates values as soon as they are computed.
- In practice, Gauss-Seidel is preferred for better performance, but both methods are widely used in numerical simulations and large-scale systems.
