**Iterative Methods**

- refining the guess
- converging to a good guess

**Stability Analysis**

```python
#JACOBI Method

#time module:
import time

import numpy as np

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x, dtype=float)
        
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
    
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new

    return x, max_iterations
```

```python
# Example usage-JACOBI

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6

max_iterations = 100

# Start time
start_time = time.time()

solution, iterations = jacobi(A, b, x0, tol, max_iterations)

# End time
end_time = time.time()

#Elaspsed time
elapsed_time = end_time - start_time
print("Jacobi:")
print(f"Elapsed time: {elapsed_time}")

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
```

```python
#GAUSS-SEIDEL Method

import numpy as np

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy().astype(float)
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k

        x = x_new

    return x, max_iterations
```

```python
# Example usage-GAUSS-SEILDEL 

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6

max_iterations = 100

# Start time
start_time = time.time()

solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)

# End time
end_time = time.time()

#Elaspsed time
elapsed_time = end_time - start_time

print("Gauss-Seidel")
print(f"Elapsed time: {elapsed_time}")

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
```

```python
import numpy as np

def is_diagonally_dominant(A):
    n = len(A)
    
    for i in range(n):
        diagonal = abs(A[i][i])
        off_diagonal_sum = sum(abs(A[i][j]) for j in range(n) if j != i)
        
        if diagonal < off_diagonal_sum:
            return False
            
    return True

# Example matrices

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)

result = is_diagonally_dominant(A)
print("Is the matrix diagonally dominant?", result)
print(A)
```
