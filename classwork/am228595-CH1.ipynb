```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'b--')
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'r-o')
plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x)*np.cos(x), 'g')
plt.subplot(2, 2, 4)
plt.plot(x, np.tan(x), 'k')
plt.show()
```
