```python
import numpy as np
import time
import matplotlib.pyplot as plt

# -----------------------------
# Naive matrix multiplication
# -----------------------------
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

# -----------------------------
# Experiment setup
# -----------------------------
sizes = [3 * (2**i) for i in range(8)]  # 3, 6, 12, 24, 48, 96, 192, 384
naive_times = []
numpy_times = []

for n in sizes:
    print(f"Running size {n}x{n}...")

    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    # --- Time naive multiplication ---
    start = time.time()
    naive_matrix_multiplication(A.tolist(), B.tolist())
    naive_times.append(time.time() - start)

    # --- Time NumPy multiplication ---
    start = time.time()
    A @ B
    numpy_times.append(time.time() - start)

# -----------------------------
# Plotting results
# -----------------------------
plt.figure(figsize=(10, 6))
plt.plot(range(1, 9), naive_times, marker='o', label="Naive O(n³)")
plt.plot(range(1, 9), numpy_times, marker='o', label="NumPy Optimized")

plt.title("Naive vs NumPy Matrix Multiplication Runtime")
plt.xlabel("Matrix Number (1–8)")
plt.ylabel("Runtime (seconds)")
plt.xticks(range(1, 9))
plt.yscale("log")  # log scale to show huge difference
plt.legend()
plt.grid(True)
plt.show()
```
